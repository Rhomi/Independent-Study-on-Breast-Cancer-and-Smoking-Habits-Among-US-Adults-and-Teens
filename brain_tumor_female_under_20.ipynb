```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
#To have plots appear right in the notebook
%matplotlib inline
```

```python
brainbysite = pd.read_excel(r'C:\Users\rhomi\Documents\Datasets\USCS_1999_2013_ASCII\brainbysite.xlsx', na_values= 0)
```

```python
brainbysite_select_col = brainbysite[['AGE','AGE_ADJUSTED_RATE','BEHAVIOR','COUNT','POPULATION','SITE','SEX','YEAR']]
```

```python
brainbysite_under_20 = brainbysite_select_col[brainbysite_select_col.AGE == "0-19"]
```

```python
brainbysite_under_20_neat = brainbysite_under_20.replace('~',np.nan)
```

```python
brainbysite_under_20_neat
```

```python
bbs_u_20 = brainbysite_under_20_neat.dropna()
```

```python
bbs_u_20_f = bbs_u_20[bbs_u_20.SEX == "Female"]
```

```python
bbs_u_20_f = bbs_u_20_f[bbs_u_20_f.YEAR != "2009-2013"]
```

```python
# bbs_u_20_f = bbs_u_20_f.set_index(bbs_u_20_f.YEAR)
```

```python
bbs_u_20_f_m = bbs_u_20_f[bbs_u_20_f.BEHAVIOR == "Malignant"]
```

```python
bbs_u_20_f_m
```

```python
ax = bbs_u_20_f_m[['SITE','COUNT']].plot(kind='bar', title ="Brain Tumor", figsize=(25, 10), legend=True, fontsize=8)
ax.set_xlabel("Site", fontsize=6)
ax.set_ylabel("Count", fontsize=12)
plt.show()
```

```python
sites = bbs_u_20_f_m['SITE']
count = bbs_u_20_f_m['COUNT']
p_df = pd.DataFrame({"Sites" : sites, "Count": count})
```

```python
fig, ax = plt.subplots(figsize=(8,6))
for label, df in p_df.groupby(p_df.index):
    df.Count.plot(kind="kde", ax=ax, label=label)
plt.legend()
```
