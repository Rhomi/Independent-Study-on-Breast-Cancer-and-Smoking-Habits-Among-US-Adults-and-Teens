```python
df_dict = {}
def find_patterns(df, range, year):
    print(range)
    for index in range:
        age = index+8
        key = "df_"+year+"_"+str(age)
        df_final = df[df['qn1'] == float(index)]
        df_final['Conversion'] = pd.Series(df_final['STUDNTID']/df_final['STUDNTID'].sum(), 
                                               index=df_final.index)
        df_dict[key] = df_final
        
    return df_dict
```
