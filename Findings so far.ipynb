```python
df_dict = {}
def find_patterns(df, range):
    print(range)
    for index in range:
        age = index+8
        key = "df_2014_"+str(age)
        df_final = df[df['qn1'] == float(index)]
        df_final['Conversion'] = pd.Series(df_final['STUDNTID']/df_final['STUDNTID'].sum(), 
                                               index=df_final.index)
        df_dict[key] = df_final
        
    return df_dict
```

```python
final_dict_11 = {}
final_dict_12 = {}
final_dict_13 = {}
final_dict_14 = {}

age_range = range(4,10)

# The index+8 in the line says that the right age is represented by index+8. Refer the codebook.
final_dict_11 = find_patterns(df_2011, age_range)
final_dict_12 = find_patterns(df_2012, age_range)
#final_dict_13 = find_patterns(df_2013, age_range)
final_dict_14 = find_patterns(df_2014, age_range)
```
